```python
sin(100)
```

```python
from a import * # 위에 임포트와 곂칠 수 있음 * 사용은 자제하는 것이 좋음 
```

#### as 
 - 모듈 import 시, alias(별명) 지정가능
 

```python
import math as m  # math 라고 다 치기 힘드니 m 이라고 내가 별명을 지워냄 

m.exp(3)
```

```python
m.cos(100)
```
